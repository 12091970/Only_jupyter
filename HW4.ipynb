Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('kc_house_data.csv', sep=',')

corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
#import seaborn as sns
#import matplotlib.pyplot as plt

#plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'],y=df['yr_built'],kind='kde');
#Распределение по году постройки с указанием наибольшей цены
#в дипазоне 1960-1980.
```

```python
sns.boxplot(x=df['price'], y=df['bedrooms'].astype('str'));  #, whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Спальни')
plt.title('Влияние количества спален на цену.');
# Наибольшая  стоимость у домов с 8 спальнями.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('kc_house_data.csv', sep=',')

sns.scatterplot(x=df['price'],y=df['yr_built'])
plt.title('Распределение цены по годам постройки')
plt.xlabel('Цена')
plt.ylabel('Год постройки')
plt.grid();

#Основная ценовая категория домов до 3 распределена в основном равномерно по всем годам,
#единичные случаи в промежутках -3-4,4-5,5-6,6-7,7-8.

```
